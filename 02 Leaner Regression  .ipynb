```python
import pandas 
from pandas import DataFrame
import matplotlib.pyplot as pit
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
x = DataFrame(data,columns=['production_budget_usd'])
y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
x.describe()
```

```python
y.describe()
```

```python
pit.figure(figsize =(13,8))
pit.title('Flim Cost VS Global Revenue')
pit.xlabel('product Budget $')
pit.ylabel('WorldWide Gross $')
pit.ylim(0,3000000000)
pit.xlim(0,450000000)

pit.scatter(x, y,alpha=0.3)
pit.show()
```

```python
regression = LinearRegression()
regression.fit(x, y)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
pit.figure(figsize =(13,8))
pit.title('Flim Cost VS Global Revenye')
pit.xlabel('product Budget $')
pit.ylabel('WorldWide Gross $')
pit.ylim(0,3000000000)
pit.xlim(0,450000000)
pit.style.use('fivethirtyeight')


pit.scatter(x, y, alpha=0.3)
pit.plot(x,regression.predict(x),color="green",linewidth=4)
pit.show()
```
